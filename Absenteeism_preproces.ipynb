```python
df[['Day', 'Month Value', 'Year']] = df['Date'].str.split('/', expand=True)

# Optionally, convert 'Day', 'Month', 'Year' to numeric if needed
df['Day'] = pd.to_numeric(df['Day'])
df['Month Value'] = pd.to_numeric(df['Month Value'])
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')

# Extract day of the week (Monday=0, Sunday=6)
df['Day of the Week'] = df['Date'].dt.dayofweek

# Map day of the week integers to weekday names
weekday_map = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
df['Weekday_Name'] = df['Day of the Week'].map(weekday_map)

# Display the updated DataFrame
print(df)
```
